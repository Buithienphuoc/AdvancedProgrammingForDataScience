```python
# Add vocab to list
vocab = {}
with open('vocab.txt', 'r') as f:
    for line in f:
        word, index = line.strip().split(':')
        vocab[word] = int(index)

len(vocab)
```

```python
cVectorizer = CountVectorizer(analyzer = "word",vocabulary = vocab) # initialised the CountVectorizer
count_features = cVectorizer.fit_transform(df['Description'])
count_features.shape
```

```python
count_features
```

```python
count_features = cVectorizer.fit_transform(df['Description']).toarray()
count_features
```

```python
def save_count_vector(count_features_param, web_index, filename):
    with open(filename, 'w') as f:
        for i in range(len(web_index)):
            print(i)
            f.write('#' + str(web_index[i]) + ',')
            for j in range(len(count_features_param[i])):
                if count_features_param[i][j] != 0:
                    f.write(str(j) + ':' + str(count_features_param[i][j]) + ',')
            f.write('\n')
    f.close()
    print('Count vector representation saved to ' + filename)


save_count_vector(count_features, df['Webindex'], 'count_vectors.txt')
```
